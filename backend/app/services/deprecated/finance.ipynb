```python
import mstarpy as ms
import pandas as pd

def get_fund_full_info(isin):
    funds = ms.Funds(isin)
    info = {}

    # Datos básicos
    info['name'] = getattr(funds, 'name', None)
    info['isin'] = getattr(funds, 'isin', None)

    # Snapshot general
    try:
        snapshot = funds.snapshot()
        info['category'] = snapshot.get('categoryName')
        info['fund_size'] = snapshot.get('fundAssets')
        if 'fundManagerList' in snapshot and snapshot['fundManagerList']:
            info['company'] = snapshot['fundManagerList'][0].get('firm')
    except Exception as e:
        info['snapshot_error'] = str(e)

    # Comisiones
    try:
        inv_fees = funds.investmentFee()
        if isinstance(inv_fees, pd.DataFrame) and not inv_fees.empty:
            for _, row in inv_fees.iterrows():
                name = str(row.get('Name', '')).lower()
                value = row.get('Value')
                if 'management' in name:
                    info['management_fee'] = value
                if 'ongoing' in name or 'expense' in name or 'ter' in name:
                    info['ongoing_charges'] = value
    except Exception as e:
        info['fees_error'] = str(e)

    # Rentabilidades
    try:
        trailing = funds.trailingReturn()
        if isinstance(trailing, dict) and 'trailingReturnList' in trailing:
            for item in trailing['trailingReturnList']:
                if item.get('timePeriod') == '1 Year' and item.get('returnType') == 'totalReturn':
                    info['return_1y'] = item.get('return')
                if item.get('timePeriod') == '3 Year' and item.get('returnType') == 'totalReturn':
                    info['return_3y'] = item.get('return')
                if item.get('timePeriod') == '5 Year' and item.get('returnType') == 'totalReturn':
                    info['return_5y'] = item.get('return')
    except Exception as e:
        info['return_error'] = str(e)

    # Riesgo
    try:
        risk = funds.riskVolatility()
        if isinstance(risk, dict):
            for item in risk.get('standardDeviation', []):
                if item.get('year') == 3:
                    info['volatility_3y'] = item.get('value')
            for item in risk.get('sharpeRatio', []):
                if item.get('year') == 3:
                    info['sharpe_3y'] = item.get('value')
    except Exception as e:
        info['risk_error'] = str(e)

    # Drawdown
    try:
        drawdown = funds.maxDrawDown()
        if isinstance(drawdown, dict) and 'maxDrawDown' in drawdown:
            for item in drawdown['maxDrawDown']:
                if item.get('timePeriod') == '3Year':
                    info['max_drawdown_3y'] = item.get('value')
    except Exception as e:
        info['drawdown_error'] = str(e)

    # ESG
    try:
        esg = funds.esgRisk()
        if isinstance(esg, dict):
            info['esg_risk_score'] = esg.get('esgRiskScore')
            info['esg_category'] = esg.get('esgRiskCategory')
    except Exception as e:
        info['esg_error'] = str(e)

    # Otros datos útiles
    try:
        style = funds.equityStyle()
        if isinstance(style, dict):
            info['equity_style'] = style.get('style')
    except Exception as e:
        info['style_error'] = str(e)

    # Convertir a DataFrame
    df = pd.DataFrame([info])
    return df

# Ejemplo de uso:
isin = "ES0110098037"
df_fondo = get_fund_full_info(isin)
print(df_fondo.T)
```

```python
import mstarpy as mstar
import pandas as pd
import inspect

def explore_mstarpy_api(isin):
    """Explore the mstarpy API structure for a single fund"""
    print(f"Exploring mstarpy API for ISIN: {isin}")
    
    try:
        # Create an instance of the Funds class
        funds = mstar.Funds(term=isin)
        
        # Examine what methods and attributes are available
        print("\nAvailable methods and attributes in Funds object:")
        for name in dir(funds):
            if not name.startswith('__'):  # Skip internal attributes
                attr = getattr(funds, name)
                if inspect.ismethod(attr):
                    print(f"- Method: {name}")
                else:
                    print(f"- Attribute: {name}")
        
        # Try to search using the search method
        print("\nTrying to search using available methods...")
        
        # Try the most likely method names
        if hasattr(funds, 'search'):
            print("Using funds.search()...")
            search_results = funds.search()
            print(f"Search results type: {type(search_results)}")
            print(search_results.head() if hasattr(search_results, 'head') else search_results)
        
        # If search doesn't work, try the first attribute that looks like a DataFrame
        for name in dir(funds):
            if not name.startswith('__') and not inspect.ismethod(getattr(funds, name)):
                try:
                    attr = getattr(funds, name)
                    if isinstance(attr, pd.DataFrame) and not attr.empty:
                        print(f"\nFound DataFrame in attribute: {name}")
                        print(attr.head())
                        
                        # If we found a DataFrame with secId, use it
                        if 'secId' in attr.columns:
                            secId = attr.iloc[0]['secId']
                            print(f"\nFound secId: {secId}")
                            
                            # Try to get profile
                            if hasattr(funds, 'get_profile'):
                                print("\nGetting profile...")
                                profile = funds.get_profile(secId)
                                print(f"Profile type: {type(profile)}")
                                if isinstance(profile, dict):
                                    print(f"Profile keys: {list(profile.keys())[:10]}")
                            
                            # Try to get fees
                            if hasattr(funds, 'get_fees'):
                                print("\nGetting fees...")
                                fees = funds.get_fees(secId)
                                print(f"Fees type: {type(fees)}")
                                if isinstance(fees, pd.DataFrame):
                                    print(fees.head())
                        break
                except Exception as e:
                    print(f"Error accessing {name}: {str(e)}")
        
        return funds
        
    except Exception as e:
        print(f"❌ Error exploring API: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# Test with one of your funds
api_exploration = explore_mstarpy_api("ES0110098037")  # BBVA Bolsa Indice Euro FI
```

```python
def test_all_mstar_fee_methods(isin):
    """Test all fee-related methods in mstarpy to find the most useful ones"""
    print(f"Testing all fee methods for ISIN: {isin}")
    
    try:
        # Create a Funds instance with the ISIN
        funds = mstar.Funds(term=isin)
        
        # Dictionary to collect fee information
        fee_data = {
            'management_fee': None,
            'ongoing_charges': None,
            'category': None,
            'company': None
        }
        
        # Method 1: feeLevel method
        print("\n1. Trying feeLevel method:")
        try:
            fee_level = funds.feeLevel()
            if isinstance(fee_level, pd.DataFrame) and not fee_level.empty:
                print(fee_level)
                for _, row in fee_level.iterrows():
                    if isinstance(row.get('Name'), str):
                        name = row.get('Name').lower()
                        if 'management' in name:
                            fee_data['management_fee'] = row.get('Value')
                        if 'ongoing' in name or 'expense' in name or 'ter' in name:
                            fee_data['ongoing_charges'] = row.get('Value')
        except Exception as e:
            print(f"Error with feeLevel: {e}")
        
        # Method 2: otherFee method
        print("\n2. Trying otherFee method:")
        try:
            other_fee = funds.otherFee()
            if isinstance(other_fee, pd.DataFrame) and not other_fee.empty:
                print(other_fee)
                # Look for fee information
                for _, row in other_fee.iterrows():
                    if isinstance(row.get('Name'), str):
                        name = row.get('Name').lower()
                        if 'management' in name:
                            fee_data['management_fee'] = row.get('Value')
                        if 'ongoing' in name or 'expense' in name or 'ter' in name:
                            fee_data['ongoing_charges'] = row.get('Value')
        except Exception as e:
            print(f"Error with otherFee: {e}")
            
        # Method 3: historicalExpenses method
        print("\n3. Trying historicalExpenses method:")
        try:
            hist_exp = funds.historicalExpenses()
            if isinstance(hist_exp, pd.DataFrame) and not hist_exp.empty:
                print(hist_exp)
                # Most recent expense ratio could be helpful
                if 'AnnualReportExpenseRatio' in hist_exp.columns:
                    fee_data['ongoing_charges'] = hist_exp['AnnualReportExpenseRatio'].iloc[0]
        except Exception as e:
            print(f"Error with historicalExpenses: {e}")
            
        # Method 4: dataPoint method might have useful info
        print("\n4. Trying dataPoint method:")
        try:
            data_point = funds.dataPoint()
            if isinstance(data_point, dict):
                print(f"Keys: {list(data_point.keys())[:10]}")
                # Try to extract fees if present
                if 'managementFee' in data_point:
                    fee_data['management_fee'] = data_point['managementFee']
                if 'ongoingCharge' in data_point:
                    fee_data['ongoing_charges'] = data_point['ongoingCharge']
                # Category info might be useful too
                if 'morningstarCategory' in data_point:
                    fee_data['category'] = data_point['morningstarCategory']
                if 'fundCompany' in data_point:
                    fee_data['company'] = data_point['fundCompany']
        except Exception as e:
            print(f"Error with dataPoint: {e}")
        
        # Method 5: Get full snapshot method for complete info
        print("\n5. Examining full snapshot dictionary:")
        try:
            snapshot = funds.snapshot()
            if 'fees' in snapshot:
                print("Fee information found in snapshot:")
                print(snapshot['fees'])
                # Try to extract from fees
                if isinstance(snapshot['fees'], dict):
                    for key, value in snapshot['fees'].items():
                        if 'management' in key.lower():
                            fee_data['management_fee'] = value
                        if 'ongoing' in key.lower() or 'expense' in key.lower():
                            fee_data['ongoing_charges'] = value
            
            # Category and company info
            if 'categoryName' in snapshot:
                fee_data['category'] = snapshot['categoryName']
                print(f"Category: {snapshot['categoryName']}")
            if 'fundManagerList' in snapshot and len(snapshot['fundManagerList']) > 0:
                fee_data['company'] = snapshot['fundManagerList'][0].get('firm')
                print(f"Company: {snapshot['fundManagerList'][0].get('firm')}")
        except Exception as e:
            print(f"Error examining snapshot: {e}")
        
        print("\nCollected fee data:")
        for key, value in fee_data.items():
            print(f"{key}: {value}")
        
        return fee_data
        
    except Exception as e:
        print(f"❌ Error retrieving data: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# Test with one of your funds
comprehensive_results = test_all_mstar_fee_methods("ES0110098037")  # BBVA Bolsa Indice Euro FI
```

```python
import mstarpy as ms
import pandas as pd

def get_full_fund_dataframe(isin):
    funds = ms.Funds(isin)
    result = {}

    # Nombre e ISIN
    result['name'] = getattr(funds, 'name', None)
    result['isin'] = getattr(funds, 'isin', None)

    # Asset Allocation (renta fija/variable/otros)
    try:
        alloc = pd.DataFrame(funds.allocationMap())
        if not alloc.empty and 'holdingType' in alloc.columns and 'weighting' in alloc.columns:
            alloc_pivot = alloc.groupby('holdingType')['weighting'].sum().reset_index()\
                .pivot_table(index=None, columns='holdingType', values='weighting')\
                .add_prefix('perc_').reset_index(drop=True)
            result.update(alloc_pivot.to_dict(orient='records')[0])
    except Exception:
        pass

    # Categoría y tamaño
    try:
        snapshot = funds.snapshot()
        result['category'] = snapshot.get('categoryName')
        result['fund_size'] = snapshot.get('fundAssets')
    except Exception:
        pass

    # Comisiones
    try:
        fees = funds.investmentFee()
        if isinstance(fees, pd.DataFrame) and not fees.empty:
            for _, row in fees.iterrows():
                name = str(row.get('Name', '')).lower()
                value = row.get('Value')
                if 'management' in name:
                    result['management_fee'] = value
                if 'ongoing' in name or 'expense' in name or 'ter' in name:
                    result['ongoing_charges'] = value
    except Exception:
        pass

    # Rentabilidades
    try:
        trailing = funds.trailingReturn()
        if isinstance(trailing, dict) and 'trailingReturnList' in trailing:
            for item in trailing['trailingReturnList']:
                if item.get('timePeriod') == '1 Year' and item.get('returnType') == 'totalReturn':
                    result['return_1y'] = item.get('return')
                if item.get('timePeriod') == '3 Year' and item.get('returnType') == 'totalReturn':
                    result['return_3y'] = item.get('return')
                if item.get('timePeriod') == '5 Year' and item.get('returnType') == 'totalReturn':
                    result['return_5y'] = item.get('return')
    except Exception:
        pass

    # Volatilidad y Sharpe
    try:
        risk = funds.riskVolatility()
        if isinstance(risk, dict):
            for item in risk.get('standardDeviation', []):
                if item.get('year') == 3:
                    result['volatility_3y'] = item.get('value')
            for item in risk.get('sharpeRatio', []):
                if item.get('year') == 3:
                    result['sharpe_3y'] = item.get('value')
    except Exception:
        pass

    # Drawdown
    try:
        drawdown = funds.maxDrawDown()
        if isinstance(drawdown, dict) and 'maxDrawDown' in drawdown:
            for item in drawdown['maxDrawDown']:
                if item.get('timePeriod') == '3Year':
                    result['max_drawdown_3y'] = item.get('value')
    except Exception:
        pass

    # ESG
    try:
        esg = funds.esgRisk()
        if isinstance(esg, dict):
            result['esg_risk_score'] = esg.get('esgRiskScore')
            result['esg_category'] = esg.get('esgRiskCategory')
    except Exception:
        pass

    # Equity Style
    try:
        style = funds.equityStyle()
        if isinstance(style, dict):
            result['equity_style'] = style.get('style')
    except Exception:
        pass

    # Sector breakdown (renta fija)
    try:
        sector = funds.sector()
        if 'FIXEDINCOME' in sector and 'fundPortfolio' in sector['FIXEDINCOME']:
            sector_df = pd.DataFrame(sector['FIXEDINCOME']['fundPortfolio'], index=[0])
            for col in sector_df.columns:
                if col != 'portfolioDate':
                    result[f'perc_sector_{col}'] = sector_df.iloc[0][col]
    except Exception:
        pass

    # Puedes añadir más métodos aquí si lo necesitas

    # Devuelve DataFrame de una fila
    return pd.DataFrame([result])

# Ejemplo de uso:
isin = "ES0110098037"
df_fondo = get_full_fund_dataframe(isin)
print(df_fondo.T)
```
