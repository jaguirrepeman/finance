# Análisis de Fondo de Inversión

Este notebook muestra cómo utilizar la clase `Fund` de `functions_fund.py` para obtener y analizar los datos de un fondo de inversión.

```python
import pandas as pd
from functions_fund import Fund
import plotly.express as px

# Descomentar la siguiente línea si functions_fund.py está en un directorio superior
# import sys
# sys.path.append('../src')
```

## 1. Cargar datos de un fondo

Introduce el ISIN del fondo que quieres analizar y crea una instancia de la clase `Fund`. Esto disparará el proceso de recolección de datos (usando el caché si está disponible).

```python
# ISIN de ejemplo (puedes cambiarlo por el que quieras)
isin_ejemplo = "ES0133337008" 

# Crear la instancia del fondo. La primera vez puede tardar un poco.
fondo = Fund(isin=isin_ejemplo, use_cache=True)

# El resultado está en el atributo fund_data
datos_fondo_df = fondo.fund_data

print(f"Datos cargados para el fondo: {datos_fondo_df['nombre'].iloc[0]}")
display(datos_fondo_df)
```

## 2. Acceder a los datos detallados

El DataFrame `datos_fondo_df` contiene una columna llamada `data` que a su vez contiene otro DataFrame con todos los detalles recopilados de Morningstar y Finect.

```python
detalles_df = datos_fondo_df['data'].iloc[0]

print("Columnas disponibles en los datos detallados:")
print(detalles_df.columns.tolist())

print("\nAlgunos datos de ejemplo:")
display(detalles_df[['isin', 'name', 'categoryName', 'ongoingCostsOtherCosts', 'rating_finect']].T)
```

## 3. Visualizar datos históricos

La columna `historical_data` contiene la serie temporal del valor liquidativo del fondo.

```python
hist_df = datos_fondo_df['historical_data'].iloc[0]

fig = px.line(hist_df, x=hist_df.index, y='Close', title=f'Evolución del Valor Liquidativo para {datos_fondo_df["nombre"].iloc[0]}')
fig.show()
```
